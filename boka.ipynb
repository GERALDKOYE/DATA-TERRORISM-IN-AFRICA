```python
def combine(d, name, seuil):
    df = attacks.copy()
    df[name] = df[d.values()]\
                        .apply(lambda x : " + ".join((str(val).strip() for val in  x[x.notnull()])) , axis = 1)
    
    df["nattackBy%s"%name] = df.groupby(name)['Date'].transform("count")
    df.loc[df[name] == "", name] =  "Other"
    df.loc[df["nattackBy%s"%name] < 100, name] = "Other"
    df["nattackBy%s"%name] = df.groupby(name)['Date'].transform("count")
    df.drop(d.values(), axis = 1, inplace  = True)
    df.drop("nattackBy%s"%name, axis = 1, inplace = True)
    return df
```

```python
df = combine(actors, "perpetrator", 100)
```
